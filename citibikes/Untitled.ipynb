```python
# iteratively reading and concatenating CSVs with list comprehension

file_name = 'data/JC-{}-citibike-tripdata.csv'
# df_list = []
# for i in range(0, 11):
#     df_list.append(pd.read_csv(file_name.format(i))

# combined_df = pd.concat(df_list)
combined_df = pd.concat([pd.read_csv(file_name.format(i)) for i in range(201801, 201813)])
```
